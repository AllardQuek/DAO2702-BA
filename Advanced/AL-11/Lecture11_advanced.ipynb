In FIFA Ultimate Teams, players with the overall rating higher than 75 are considered to be "gold" players. Players with ratings between 65 and 75 are "silver" players, and those lower than 65 are "bronze" players. Suppose EA is providing three types of player packs:

- Legendary (two players): each player has 60% chance to be gold, 30% chance to be silver, and 10% chance to be bronze. 
- Rare (five players): each player has 20% chance to be gold, 30% chance to be silver,and 50% chance to be bronze. 
- Common (eight players): each player has 10% to be gold, 20% to be silver, and 70% to be bronze. 

The game is played following the rules below:
1. A certain amount of coins are given to the player;
2. In each round of the game, the game player decides whether to buy a Legendary pack (2500 coins), a Rare pack(1200 coins), or a Common pack (400 coin). 
3. Display the random soccer players included in the pack
4. The game continues until the game player use up all coins or the remaining coins are not enough to even buy a Common pack. 

Our ultimate goal is to create a function `player_pack(coins)`, with `coins` specifying the coins initially given to the player. Similar to our previous exercises, it is easier to break this goal into a number of functions.


Firstly we write a function named `random_level()` that randomly returns a dictionary in the format of `{'gold': n1, 'silver': n2, 'bronze': n3}`, where integers `n1`, `n2`, and `n3` are randomly generated while assuming the probability of having a gold player is `pg`, and the probability of having a silver player is `ps`. The summation of `n1`, `n2`, and `n3` is `n`. 

```python
import numpy as np

def random_level(pg, ps, n):
    """
    The function random_level returns a dict type object
    that indicates the randomly generated number of gold, 
    silver, and bronze players.
    
    Input Arguments
    ---------------
    pg: the probability of drawing a gold level player
    ps: the probability of drawing a silver level player
    n: the total number of players
    
    Return
    ------
    A dict type object in the format of 
    {'gold': n1, 'silver': n2, 'bronze': n3}
    where n1, n2, and n3 represent the numbers of gold, 
    silver, and bronze level players, respectively. The 
    sum of n1, n2, and n3 should be the same as the given
    integer n.
    """
    
    sample = np.random.choice(['gold', 'silver', 'bronze'], 
                              p=[pg, ps, 1-pg-ps], size=n)
    
    return {'gold': (sample == 'gold').sum(),
            'silver': (sample == 'silver').sum(),
            'bronze': (sample == 'bronze').sum()}
    

random_level(0.2, 0.3, 5)
```

Then we define a function `sample_player()` to randomly select a given number of players of a certain level from the dataset. 

```python
def sample_player(level, n, dataset):
    """
    The function draw_player randomly draws n players belong 
    to the given level from the dataset 
    """
    
    if level == 'gold':
        subset = dataset.loc[dataset['Overall'] >= 75]
    elif level == 'silver':
        subset = dataset.loc[(dataset['Overall'] >= 65) & (dataset['Overall'] < 75)]
    else:
        subset = dataset.loc[dataset['Overall'] < 65]
    
    return subset.sample(n, replace=True)

players = sample_player('gold', 5, fifa)
print_players(players)
```

Lastly, we use functions defined above to build the program for purchasing player packs. 

```python
def player_packs(coins):
    """
    The player_packs function minics the player packs of the FIFA2021 Ultimate Teams.
    Rules of the game:
    The player is given a number of coins (as the input argument) to play the game.
    1. In each round, the player decides whether to buy a 'Legendary' pack (2500 
       coins), a 'Rare' pack (1200 coins), or a 'Common' pack (400 coins). 
    2. A Legendary pack generates 2 players, with 60% gold, 30% silver, and 10% 
       bronze.
    3. A Rare pack generates 5 players, with 20% gold, 30% silver, and 50% bronze.
    4. A Common pack generates 8 players, with 10% gold, 20% silver, and 70%
       bronze.
    5. The game ends when the player used up all coins, or does not have enough coins
       to purchase even a Common pack.
    """
    
    prices = {'L': 2500,    # Cost of the Legendary packs
              'R': 1200,    # Cost of the Rare packs
              'C': 400}     # Cost of the Common packs
    
    while coins >= prices['C'] :
        
        print('You have {0} coins'.format(coins))
        pack = input('Choose a pack: L for Legendary, R for Rare, C for Common: ').upper()
        
        if pack not in prices:
            print('Invalid input, please try again. ')
            print('\n')
            continue
        
        if coins - prices[pack] < 0:
            print('You cannot afford it, please try again. ')
            print('\n')
            continue
        else:
            coins -= prices[pack]
        
        if pack == 'L':
            pg, ps, n = 0.6, 0.3, 2
        elif pack == 'R':
            pg, ps, n = 0.2, 0.3, 5
        else:
            pg, ps, n = 0.1, 0.2, 8
        
        print('\n')
        num_player = random_level(pg, ps, n)
        
        for level, count in num_player.items():
            if count > 0:
                print_players(sample_player(level, count, fifa))
        
player_packs(3000)      
```
