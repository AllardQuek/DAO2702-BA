The attached dataset provides the information on employees of a big tech company in Seattle, in term of:

- title: title of positions
- salary: gross salary
- total_exper: the total number of years of working experience
- company_exper: the number of years working for the company
- gender: gender of the employee
- race: race of the employee
- education: the highest degree the employee holds


Please use data visualization to assess the influence of some factors on employees' salaries.
```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('hr_seattle.csv')
df.head()
```

```python
# Plot each of the fields against salary

for field in df.columns:
    if field != 'salary':
        plt.figure(figsize=(10, 5))
        sns.boxplot(x=field, y='salary', data=df)
        plt.xticks(rotation=90)
        plt.show()
```

```python
# Plot bar plot of title v.s. salary, with one bar for each gender
sns.barplot(x="title", y="salary", hue="gender", data=df)
plt.xticks(rotation=90)
```

```python
# Plot bar plot of title v.s. salary, with one bar for each race
sns.barplot(x="title", y="salary", hue="race", data=df)
plt.xticks(rotation=90)
```
