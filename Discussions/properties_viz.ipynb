```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('properties3.csv')
df.head()
```

```python
# Type of property v.s. price
df.groupby('type').price.mean().plot(kind='bar')
plt.xlabel('Type of property')
plt.ylabel('Average price')
plt.show()

```

```python
# Plot distribution of type and sale
sns.countplot(x='type', data=df, hue='sale')
plt.xlabel('Type of property')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the relationship between price and area
df.plot(kind='scatter', x='area', y='price')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

```python
# Plot relationship between area and price, differentiating colours with the type of property
sns.lmplot(x='area', y='price', data=df, hue='type', fit_reg=False)
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

```python
# Plot relationship between area and price, differentiating colours with the type of sale
sns.lmplot(x='area', y='price', data=df, hue='sale', fit_reg=False)
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```
